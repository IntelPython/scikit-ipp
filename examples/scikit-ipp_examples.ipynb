# scikit-ipp _(skipp)_
### Introductory examples for scikit-ipp

## Gaussian filter

```python
from skimage.data import astronaut
from skipp.skipp.filters import gaussian as skipp_gaussian
from skimage.filters import gaussian as skimage_gaussian
from scipy.ndimage.filters import gaussian_filter as scipy_gaussian
import numpy as np
import matplotlib.pyplot as plt
```

```python
from skimage._shared.testing import (assert_equal, assert_array_equal, 
                                     assert_almost_equal, assert_allclose)
```

```python
def showImagesGRAY(original, result):
    fig, axes = plt.subplots(1, 2, figsize=(8, 4))
    ax = axes.ravel()

    ax[0].imshow(original, cmap='gray')
    ax[0].set_title("original")
    ax[1].imshow(result, cmap='gray')
    ax[1].set_title("result")
    
def showImages(original, result):
    fig, axes = plt.subplots(1, 2, figsize=(8, 4))
    ax = axes.ravel()

    ax[0].imshow(original)
    ax[0].set_title("original")
    ax[1].imshow(result)
    ax[1].set_title("result")

def imageInfo(image):
    print("Dimensions: ", image.ndim, 
          "\nStrides: ", image.strides, 
          "\nShape: ", image.shape, 
          "\nDtype: ", image.dtype)
```

### Example with float32 image

```python
someImage = np.arange(81, dtype =np.float32).reshape((9,9))
someImage
```

```python
filtered_skipp_gaussian = skipp_gaussian(someImage, output=np.float32, sigma=10.25)
someImage
```

```python
filtered_skimage_gaussian = skimage_gaussian(someImage, output=np.float32, sigma=10.25)
showImagesGRAY(filtered_skipp_gaussian, filtered_skimage_gaussian)
```

```python
np.allclose(filtered_skipp_gaussian, filtered_skimage_gaussian, atol=1e-4)
```

```python
assert assert_almost_equal(filtered_skimage_gaussian, filtered_skipp_gaussian)
```

```python
#%timeit -n 10 -r 2 filtered_ScikitIPP = GaussianFilterBorderFuncFLOAT32(bigImage, sigma=10.25, output=destination, kernelSize=1023)
%timeit filtered_skipp_gaussian = skipp_gaussian(someImage, output = np.arange(81,dtype=np.uint8).reshape((9,9)), sigma=10.25)
#%timeit -n 10 -r 2 filtered_skimage = gaussian(bigImage, sigma=10.25)
%timeit filtered_skimage_gaussian = skimage_gaussian(someImage, output=np.uint8, sigma=10.25)
```

### Example with uint8 image

```python
someImage = np.arange(25, dtype =np.uint8).reshape((5,5))
```

```python
filtered_skipp_gaussian = skipp_gaussian(someImage, output=np.uint8, sigma=10.25, mode = "nearest")
```

```python
filtered_skimage_gaussian = skimage_gaussian(someImage, output=np.uint8, sigma=10.25, mode = "nearest")
showImagesGRAY(filtered_skipp_gaussian, filtered_skimage_gaussian)
```

```python
#%timeit -n 10 -r 2 filtered_skipp_gaussian = skipp_gaussian(someImage, output=np.uint8, sigma=10.25,mode = "nearest")
%timeit filtered_skipp_gaussian = skipp_gaussian(someImage, output=np.uint8, sigma=10.25,mode = "nearest")
#%timeit -n 10 -r 2 filtered_skimage_gaussian = skimage_gaussian(someImage, output=np.uint8, sigma=10.25, mode = "nearest")
%timeit filtered_skimage_gaussian = skimage_gaussian(someImage, output=np.uint8, sigma=10.25, mode = "nearest")
```

```python
np.allclose(filtered_skipp_gaussian, filtered_skimage_gaussian, atol=1e+0)
```

```python
from skimage._shared.testing import (assert_equal, assert_array_equal, 
                                     assert_almost_equal, assert_allclose)
```

```python
assert assert_equal(filtered_skipp_gaussian, filtered_skimage_gaussian)
```

```python
assert assert_almost_equal(filtered_skipp_gaussian, filtered_skimage_gaussian)
```

```python
assert assert_allclose(filtered_skipp_gaussian, filtered_skimage_gaussian)
```
